```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
matplotlib.style.use('ggplot') # Make it look pretty
```

```python
df = pd.read_csv('./Datasets/wheat.data')
```

```python
df.columns
```

```python
df.asymmetry.plot.hist(title='Asymmetry', bins=6)
```

```python
plt.show()
```

```python
df.asymmetry.plot.hist(title='Asymmetry', bins=10)
plt.show()
```

```python
df[df.wheat_type=='rosa'].asymmetry.plot.hist(alpha=0.4)
df[df.wheat_type=='kama'].asymmetry.plot.hist(alpha=0.4)
df[df.wheat_type=='canadian'].asymmetry.plot.hist(alpha=0.4)
plt.show()
```

```python
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')

student_dataset = pd.read_csv('./Datasets/students.data', index_col=0)

my_series = student_dataset.G3
my_dataframe = student_dataset[['G3', 'G2', 'G1']]

my_series.plot.hist(alpha=0.5)
my_dataframe.plot.hist(alpha=0.5)
```

```python
# This bit missing from discussion
plt.show()
```

```python
# Probabilities per bin
# Note this changes value in Frequency (y axis)
my_series = student_dataset.G3
my_dataframe = student_dataset[['G3', 'G2', 'G1']]

my_series.plot.hist(normed=True, alpha=0.5)
my_dataframe.plot.hist(normed=True, alpha=0.5)
```
