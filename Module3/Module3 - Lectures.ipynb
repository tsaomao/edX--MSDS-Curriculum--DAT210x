# 3D Scatter Plots
[Lecture](https://courses.edx.org/courses/course-v1:Microsoft+DAT210x+4T2017/courseware/38ed0930-5f2a-0f27-ce1d-33744fd89c2f/f3585b48-ab18-3729-0b7a-924d7087a9ea/)

Pandas dataframes don't natively support 3D plotting, so instead you have to write out the connectivity/mapping yourself. Per the lecture:

```python
# Magic command, works inside jupyter notebooks
# This includes an interactive control/renderer and does not require plt.show()
%matplotlib notebook

# This one below dispenses with having to use plt.show() and instead immediately displays plots.
# %matplotlib inline

# This option pops out the plot into a separate window rather than within the notebook.
# %matplotlib gtk
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
matplotlib.style.use('ggplot') # Pretty
```

```python
df = pd.read_csv('./Datasets/concrete.csv', index_col=0)
```

```python
df.columns = ['Cement', 'Slag', 'Ash', 'Water', 'SP', 'Coarse', 'Fine', 'SLUMP', 'FLOW', 'Strength']
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Cement Content')
ax.set_ylabel('Water Content')
ax.set_zlabel('Concrete Strength')
ax.scatter(df.Cement, df.Water, df.Strength, c='r', marker='o')

plt.suptitle('Cement x H2O x Concrete Strength')

# Note the upward trend of Strength when there's a combination of increased Water
# AND increased Cement
# This reinforces the importance at collecting and looking at multiple features
# for every observation, not just the ones you suspect, but the (relevant)
# ones you can collect
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Cement Content')
ax.set_ylabel('Slag Content')
ax.set_zlabel('Concrete Strength')
ax.scatter(df.Cement, df.Slag, df.Strength, c='g', marker='.')

plt.suptitle('Cement x Slag x Concrete Strength')

# Note this shows that a lot of these observations were collected with a Slag Content of 0
# This is not incredibly helpful for us because of lack of Gaussian distribution and low sampling
# for slag content of between 25 and 75
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Cement Content')
ax.set_ylabel('Fly Ash Content')
ax.set_zlabel('Concrete Strength')
ax.scatter(df.Cement, df.Ash, df.Strength, c='b', marker='^')

plt.suptitle('Cement x Fly Ash x Concrete Strength')
# Note the plane in 3 dimensions (looking on diagonal between Cement and Fly Ash, and climbing Concrete Strength)
# Probably would want to collect more samples in the missing data areas.
```
