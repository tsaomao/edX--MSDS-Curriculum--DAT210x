```python
# Uses the Image module (PIL)
from scipy import misc
import os

pathstr = '/Users/malcolmgin/Dropbox/MSDS/DAT210x-working/edX--MSDS-Curriculum--DAT210x/Module4/Datasets/ALOI/32'
directory = os.fsencode(pathstr)

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".png"): 
        # Load the image up
        img = misc.imread(pathstr + '/' + filename)
        samples.append(img.flatten().reshape(27648, 1)[:, 0])
        colors.append('b')
        continue
```

Convert `samples` to a DataFrame named `df`:

```python
df = pd.DataFrame(samples)
```

Import any necessary libraries to perform Isomap here, reduce `df` down to three components and using `K=6` for your neighborhood size:

```python
from sklearn import manifold
iso = manifold.Isomap(n_neighbors=6, n_components=3)
T = iso.fit_transform(df)
```

Create a 2D Scatter plot to graph your manifold. You can use either `'o'` or `'.'` as your marker. Graph the first two isomap components:

```python
num_images, num_pixels = df.shape
num_pixels = int(math.sqrt(num_pixels))

fig = plt.figure()
    
ax = fig.add_subplot(111)
ax.set_title('First Two')
ax.set_xlabel('x')
ax.set_ylabel('y')
    
x_size = (max(T[:,0]) - min(T[:,0])) * 0.08
y_size = (max(T[:,1]) - min(T[:,1])) * 0.08
    
ax.scatter(T[:,0],T[:,1], marker='o',alpha=1)
```

Chart a 3D Scatter plot to graph your manifold. You can use either `'o'` or `'.'` as your marker:

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.style.use('ggplot')

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(T[:, 0], T[:, 1], T[:, 2], c = 'red', marker='o')

ax.set_title('Isomap 3D plot')
ax.set_xlabel('T2:0')
ax.set_ylabel('T2:1')
ax.set_zlabel('T2:2')

plt.show()
```
