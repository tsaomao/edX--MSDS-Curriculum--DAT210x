Import and alias Pandas:

```python
import pandas as pd
```

Load up the table from the link, and extract the dataset out of it. If you're having issues with this, look carefully at the sample code provided in the reading:

```python
dataframes = pd.read_html('http://www.espn.com/nhl/statistics/player/_/stat/points/sort/points/year/2015/seasontype/2')
df1 = dataframes[0]
df1
```

Next up, rename the columns so that they are _similar_ to the column definitions provided to you on the website. Be careful and don't accidentally use any column names twice. If a column uses special characters, you can replace them with regular characters to make it easier to work with:

```python
col_names = ['RK', 'Player', 'Team', 'Games-Played', 'Goals', 'Assists', 'Points', 'Plus-Minus', 'Penalty-Minutes', 'Points-per-Game', 'Shots-on-Goal', 'Shooting-Percentage', 'Game-Winning-Goals', 'Power-Play-Goals', 'Power-Play-Assists', 'Short-Handed-Goals', 'Short-Handed-Assists']
df1.columns = col_names
df1
```

Get rid of any row that has at least 4 NANs in it. That is, any rows that do not contain player points statistics:

```python
df1_filt = df1.loc[df1.isnull().sum(axis=1) < 4, :]
df1_filt
```

At this point, look through your dataset by printing it. There probably still are some erroneous rows in there. What indexing command(s) will you use to select all rows EXCEPT those rows?

```python
df1_filt2 = df1_filt.loc[df1_filt.loc[:, 'Player'] != 'PLAYER', :]
df1_filt2
```

Get rid of the 'RK' column:

```python
df1_filt3 = df1_filt2.loc[:, ['Player', 'Team', 'Games-Played', 'Goals', 'Assists', 'Points', 'Plus-Minus', 'Penalty-Minutes', 'Points-per-Game', 'Shots-on-Goal', 'Shooting-Percentage', 'Game-Winning-Goals', 'Power-Play-Goals', 'Power-Play-Assists', 'Short-Handed-Goals', 'Short-Handed-Assists']]
df1_filt3
```

Make sure there are no holes in your index by resetting it. There is an example of this in the reading material. By the way, drop the original index.

```python
df1_filt3 = df1_filt3.reset_index(drop=True)
df1_filt3
```

Check the data type of all columns, and ensure those that should be numeric are numeric.

```python
df1_filt3.loc[:, 'Games-Played'] = pd.to_numeric(df1_filt3.loc[:, 'Games-Played'], errors='coerce')
df1_filt3.loc[:, 'Goals'] = pd.to_numeric(df1_filt3.loc[:, 'Goals'], errors='coerce')
df1_filt3.loc[:, 'Assists'] = pd.to_numeric(df1_filt3.loc[:, 'Assists'], errors='coerce')
df1_filt3.loc[:, 'Points'] = pd.to_numeric(df1_filt3.loc[:, 'Points'], errors='coerce')
df1_filt3.loc[:, 'Plus-Minus'] = pd.to_numeric(df1_filt3.loc[:, 'Plus-Minus'], errors='coerce')
df1_filt3.loc[:, 'Penalty-Minutes'] = pd.to_numeric(df1_filt3.loc[:, 'Penalty-Minutes'], errors='coerce')
df1_filt3.loc[:, 'Points-per-Game'] = pd.to_numeric(df1_filt3.loc[:, 'Points-per-Game'], errors='coerce')
df1_filt3.loc[:, 'Shots-on-Goal'] = pd.to_numeric(df1_filt3.loc[:, 'Shots-on-Goal'], errors='coerce')
df1_filt3.loc[:, 'Shooting-Percentage'] = pd.to_numeric(df1_filt3.loc[:, 'Shooting-Percentage'], errors='coerce')
df1_filt3.loc[:, 'Game-Winning-Goals'] = pd.to_numeric(df1_filt3.loc[:, 'Game-Winning-Goals'], errors='coerce')
df1_filt3.loc[:, 'Power-Play-Goals'] = pd.to_numeric(df1_filt3.loc[:, 'Power-Play-Goals'], errors='coerce')
df1_filt3.loc[:, 'Power-Play-Assists'] = pd.to_numeric(df1_filt3.loc[:, 'Power-Play-Assists'], errors='coerce')
df1_filt3.loc[:, 'Short-Handed-Goals'] = pd.to_numeric(df1_filt3.loc[:, 'Short-Handed-Goals'], errors='coerce')
df1_filt3.loc[:, 'Short-Handed-Assists'] = pd.to_numeric(df1_filt3.loc[:, 'Short-Handed-Assists'], errors='coerce')
df1_filt3.dtypes
#df1_filt3
```
