# Feature Representation
([lecture](https://courses.edx.org/courses/course-v1:Microsoft+DAT210x+4T2017/courseware/b7aa9093-0ee1-9e62-4188-6f9d25a2ce85/ce1c9548-56c6-5421-5541-94aec8434947/?activate_block_id=block-v1%3AMicrosoft%2BDAT210x%2B4T2017%2Btype%40sequential%2Bblock%40ce1c9548-56c6-5421-5541-94aec8434947))

```python
import pandas as pd
```

```python
ordered_satisfaction = ['Very Unhappy', 'Unhappy', 'Neutral', 'Happy', 'Very Happy']
df = pd.DataFrame({'satisfaction': ['Mad', 'Happy', 'Unhappy', 'Neutral']})
df
```

```python
df['satisfaction2'] = df.satisfaction.astype(dtype='category',
                                            ordered = True,
                                            categories = ordered_satisfaction
                                            )
```

```python
df
```

Note that the undefined (in ordered_satisfaction) value, 'Mad' created a NaN in the satisfaction2 column.

```python
df.dtypes
```

Pandas implements a category as regular Python list.

Convert to numeric fashion, with the category codes from the index in ordered_satisfaction.

```python
df.satisfaction2 = df.satisfaction2.cat.codes
df.satisfaction2
```

```python
df
```

Note the NaN value is encoded now as -1.

```python
df = pd.DataFrame({
    'vertebrates': [
        'Bird',
        'Bird',
        'Mammal',
        'Fish',
        'Amphibian',
        'Reptile',
        'Mammal'
    ]
})
df
```

```python
df['vertebrates2'] = df.vertebrates.astype("category").cat.codes
df
```

(Can erroneously/systematically introduce data relationships. Here, e.g., it seems like a Bird is 2 integer value difference from a Mammal, even though this is just related to the Alphabetic sort used to set up the typed category codes.)

```python
df2 = pd.get_dummies(df, columns=['vertebrates'])
df2
```

(This is potentially more nuanced because it doesn't introduce the implied relationship that using `.astype("category").cat.codes` did above. It takes up more "space" or more features, but avoids introducing implied relationships. Perhaps think twice for data sets that have **many** categorical features.)

# Feature Representation - Continued

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "Authman ran faster than Harry because Authman is real and Harry is fictional.",
    "Authman and Harry ran faster then faster and faster.",
    "W00t."
]
```

```python
bow = CountVectorizer()
X = bow.fit_transform(corpus) # Produces a Sparse Matrix
X
```

```python
bow.get_feature_names()
```

```python
X.toarray()
```
